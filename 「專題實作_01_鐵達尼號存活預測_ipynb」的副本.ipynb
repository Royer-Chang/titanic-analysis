專題實作 #01: 鐵達尼號存活預測

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
# 繪製票價（Fare）與年齡（Age）之間的散佈圖
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fare', y='Age', hue='Survived', data=df)
plt.title('Fare vs Age by Survival')
plt.show()
```

```python
# 繪製年齡（Age）與生存（Survived）之間的箱形圖
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survived')
plt.show()
```

```python
# 繪製票價（Fare）與生存（Survived）之間的小提琴圖
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survived')
plt.show()
```

```python
# 繪製不同艙等（Pclass）與生存（Survived）之間的條形圖
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Pclass vs Survived')
plt.show()
```

```python
# 使用 FacetGrid 繪製年齡（Age）與生存（Survived）之間的直方圖
g = sns.FacetGrid(df, col='Survived')
g.map(plt.hist, 'Age', bins=20)
plt.show()
```

```python
# 繪製票價（Fare）與年齡（Age）之間的散佈圖
plt.figure(figsize=(10, 6))
plt.scatter(df['Fare'], df['Age'], c=df['Survived'], cmap='viridis')
plt.colorbar(label='Survived')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Fare vs Age by Survival')
plt.show()
```

```python
# 繪製年齡（Age）與生存（Survived）之間的箱形圖
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Survived')
plt.title('Age vs Survived')
plt.suptitle('')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# 繪製不同艙等（Pclass）與生存（Survived）之間的條形圖
plt.figure(figsize=(10, 6))
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title('Pclass vs Survived')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```
